## stop words

```python
from nltk.corpus import stopwords
```

```python
words = stopwords.words('English')
words
```

```python
x= 'Hello world, i am a programmer!! and you are ?'
y = word_tokenize(x)
stop = []
for i in y:
    if i not in words:
        stop.append(i)
stop
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
```

```python
a = 'been had done languages cities mice playing there better good doing eaten ate travelling movies shining'

# Tokenize the sentence and convert to lowercase
b = word_tokenize(a.lower())
```

```python
# Initialize a list to store lemmatized tokens
lemmatized_tokens = []

# Lemmatize each token and append to the list
for token in b:
    lemmatized_token = lemma.lemmatize(token)
    lemmatized_tokens.append(lemmatized_token)

# Print the lemmatized tokens
print(lemmatized_tokens)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents = ['This is a first document',
            'This is a first document',
            'Hello Third document']

tfidf = TfidfVectorizer()
matrix = tfidf.fit_transform(documents)
print(matrix.toarray())
print(tfidf.get_feature_names_out())

```
